### Import libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torchsummary import summary

import numpy as np
import matplotlib.pyplot as plt
import cv2

import copy
```

### Define problem params

```python
OPTIMIZER = 'adam'
lr = 1                # small LR requires much more iterations, large LR may not converge enough
alpha = 1e-7             # content weight
beta = 1e-5              # style weight
iter_cnt = 8000          # number of iterations
mstone_every = 800       # show result every "mstone_every" iters

# Image Files
content_img_path = 'images/content_s.png'
style_img_path = 'images/style_01.png'

VGG19_path = 'models/vgg19-d01eb7cb.pth'
```

### Check and use CUDA is available

```python
device = ("cuda" if torch.cuda.is_available() else "cpu")
print(device + " is used.")
```

### Load, show, save images

```python
# Load image file
def load_image(path):
    # Images loaded as BGR
    img = cv2.imread(path)    
    return img

# Show image
def show_image(img):
    # Convert from BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # imshow() only accepts float [0,1] or int [0,255]
    img = np.array(img/255).clip(0,1)
    
    plt.figure(figsize=(8, 4))
    plt.imshow(img)
    plt.show()
    
# Save Image
def save_image(img, name):
    img = img.clip(0, 255)
    cv2.imwrite('images\out\saved_' + str(name) + '.png', img)
```

### Image2Tensor and Tensor2Image

```python
# Preprocessing
def img_to_tensor(img):
    H, W, C = img.shape
    i2t_trans = transforms.Compose([
        transforms.ToPILImage(),
        transforms.ToTensor()
    ])
    
    # Normalize
    normalize_t = transforms.Normalize([103.939, 116.779, 123.68],[1,1,1])
    tensor = normalize_t(i2t_trans(img)*255)
    
    tensor = tensor.unsqueeze(dim=0)
    return tensor

def tensor_to_img(tensor):
    # Inverse norm
    t2i_trans = transforms.Compose([
        transforms.Normalize([-103.939, -116.779, -123.68],[1,1,1])])
    
    tensor = tensor.squeeze()
    img = t2i_trans(tensor)
    img = img.cpu().numpy()
    
    img = img.transpose(1, 2, 0)
    return img
```

### Load example

```python
# Load Images
content_img = load_image(content_img_path)
style_img = load_image(style_img_path)    

# Show Images
show_image(content_img)
show_image(style_img)
```

### Load VGG19 model

```python
#vgg = models.vgg19(pretrained=True) # Need to update jupyter and ipywidgets, for know load weights from file
vgg = models.vgg19(pretrained = False)

# Load weights
vgg.load_state_dict(torch.load(VGG19_path), strict=False)

model = copy.deepcopy(vgg.features)
model.to(device)

for param in model.parameters():
    param.requires_grad = False
```

```python
vgg
```

### Define Loss

```python
mse_loss = torch.nn.MSELoss()
def gram(tensor):
    B, C, H, W = tensor.shape
    x = tensor.view(C, H * W)
    return torch.mm(x, x.t())

def content_loss(g, c):
    loss = mse_loss(g, c)
    return loss
    
def style_loss(g, s):
    c1,c2 = g.shape
    loss = mse_loss(g, s)
    return loss / (c1**2)
```

### VGG Forward Pass

```python
def forward(model, tensor):
    layers = {
        '3':   'relu1_2',  
        '8':   'relu2_2',
        '17' : 'relu3_3',
        '26' : 'relu4_3',
        '35' : 'relu5_3',
        
        '24' : 'relu4_2'
    }
    
    # Get features
    features = {}
    x = tensor
    for name, layer in model._modules.items():
        x = layer(x)
        if name in layers:
            if (name=='24'):   
                features[layers[name]] = x
            elif (name=='31'):
                features[layers[name]] = x
            else:
                b, c, h, w = x.shape
                features[layers[name]] = gram(x) / (h*w)
                
            # Terminate forward pass
            if (name == '35'):
                break
            
    return features
```

### Define initial input image

```python
def init_img(content_tensor):
    _, C, H, W = content_tensor.shape
    tensor = torch.randn(C, H, W).mul(0.001).unsqueeze(0)    
    return tensor

# Convert image to Tensor
content_tensor = img_to_tensor(content_img).to(device)
style_tensor = img_to_tensor(style_img).to(device)
g = init_img(content_tensor)
g = g.to(device).requires_grad_(True)
```

### Transfer style

```python
def transfer_style(iter_cnt, optimizer):     
    content_layers = ['relu4_2']
    content_weights = {'relu4_2': 1 }
    style_layers = ['relu1_2', 'relu2_2', 'relu3_3', 'relu4_3', 'relu5_3']
    style_weights = {'relu1_2': 0.2, 'relu2_2': 0.2, 'relu3_3': 0.2, 'relu4_3': 0.2, 'relu5_3': 0.2}
    content = forward(model, content_tensor)
    style = forward(model, style_tensor)
    
    cnt = [0]
    while cnt[0] < iter_cnt:
        def closure():
            optimizer.zero_grad()
            # Forward pass
            gram = forward(model, g)
            # Compute loss
            c_loss=0
            s_loss=0
            for j in content_layers:
                c_loss += content_weights[j] * content_loss(gram[j], content[j])
            for j in style_layers:
                s_loss += style_weights[j] * style_loss(gram[j], style[j])
            
            total_loss =  alpha * c_loss + beta * s_loss

            # Backprop 
            total_loss.backward(retain_graph=True)
            
            # Show progress
            cnt[0] += 1
            if ((cnt[0] % mstone_every) == 1) or (cnt[0] == iter_cnt):
                print("Style Loss = {} Content Loss= {} Total Loss : {}".format(s_loss.item(), c_loss.item(), total_loss.item()))
                g_cpu = tensor_to_img(g.clone().detach())
                show_image(g_cpu)
                save_image(g_cpu, cnt[0]-1)
                plt.show()
            
            return (total_loss)
        
        # optimizer step https://pytorch.org/docs/stable/optim.html#optimizer-step-closure
        optimizer.step(closure)

    return g
```

### Define optimizer and run style transfer

```python
if (OPTIMIZER=='adam'):
    optimizer = optim.Adam([g], lr=lr)
else:
    raise ("Optimizer not defined")
    
out = transfer_style(iter_cnt, optimizer)
```

### Final result

```python
show_image(tensor_to_img(g.clone().detach()))
show_image(content_img)
save_image(tensor_to_img(g.clone().detach()), "transferred")
```
